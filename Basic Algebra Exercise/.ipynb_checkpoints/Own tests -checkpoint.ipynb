```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Hidden function
#y = 1 + 2 / 1 + x**2
def calc_y(x):
    return (1 + 1 / (-1 + x**2)) + 2*3
```

```python
x = [x for x in range(3, 9)]
numbers = [(num , calc_y(num))for num in x]
points = np.array(numbers)
print(points)
x = points[:, 0]
y = points[:, 1]
coefficients = np.polyfit(x, y, len(points) - 1)
poly = np.poly1d(coefficients)
plot_x = np.linspace(np.min(x), np.max(x), 1000)
plot_y = poly(plot_x)
plt.plot(plot_x, plot_y, c = "green")
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def interpolate_polynomial(points, degree, min_x, max_x, title = False):
    """
    Interpolates a polynomial of the specified degree through the given points and plots it
    points - a list of points (x, y) to plot
    degree - the polynomial degree
    min_x, max_x - range of x values used to plot the interpolating polynomial
    """
    x = points[:, 0]
    y = points[:, 1]
    n_points = 1000
    coefficients = np.polyfit(x, y, deg = degree)
    poly = np.poly1d(coefficients)
    x_plot = np.linspace(min_x, max_x, n_points)
    y_plot = poly(x_plot)
    plt.scatter(x, y)
    plt.plot(x_plot, y_plot, c = "g")
    if title:
        plt.title(f"{title}", loc = 'left')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
interpolate_polynomial(points, len(points) - 1, np.min(points[:, 0]), np.max(points[:, 0]), "Polynomial interpolation")
```

```python
interpolate_polynomial(points, 5, np.min(points[:, 0]), np.max(points[:, 0]), "Changing the degree to 5")
interpolate_polynomial(points, 7, np.min(points[:, 0]), np.max(points[:, 0]), "Changing the degree to 7")
interpolate_polynomial(points, 10, np.min(points[:, 0]), np.max(points[:, 0]), "Changing the degree to 10")
```

```python
interpolate_polynomial(points, 5, -10, 10, "Further expansion")
interpolate_polynomial(points, 7, -10, 10, "Further expansion")
interpolate_polynomial(points, 10, -10, 10, "Further expansion")
```

```python
interpolate_polynomial(points, len(points) - 1, -5, 15, "Further expansion")
interpolate_polynomial(points, len(points) + 1, -5, 15, "Further expansion")
```

```python
x_nonpoly = np.linspace(-10, 10, 100)
y_nonpoly = (1 + 1 / (-1 + x_nonpoly**2)) + 2*3
plt.plot(x_nonpoly, y_nonpoly)
plt.scatter(points[:, 0], points[:, 1], c = "r")
#plt.gca().set_aspect("equal")
plt.show()
```

### Complex nums

```python
def plot_complex_numbers(numbers, colors):
    """
    Plots the given complex numbers as radius vectors in the 2D space
    """
    number_of_zeros = [0] * len(numbers)
    real_parts = [num.real for num in numbers]
    imag_parts = [num.imag for num in numbers]
    plt.quiver(number_of_zeros, number_of_zeros, real_parts, imag_parts, angles = "xy", scale_units = "xy", scale = 1, color = colors)
    plt.xticks(range(-10, 10))
    plt.yticks(range(-10, 10))
#     plt.gca().set_aspect("equal") # Set aspect if needed 
    plt.scatter(real_parts, imag_parts) #show the complex nums if needed
    plt.show()
# plot_complex_numbers([2 + 3j, -2 - 1j, -3, 2j], ["green", "red", "blue", "orange"])
    plt.show()
plot_complex_numbers([2 + 3j, 4 + 1j, -2 - 1j, 2 + 1j], ["green", "red", "blue", "yellow"])
```

```python
plot_complex_numbers([2 + 1j, 2 - 1j], ["green", "red"])
```

```python
plot_complex_numbers([-2 + 1j,  -2 -1j], ["green", "red"])
```

```python
plot_complex_numbers([2 + 0j, 0 + 2j, -2 + 0j, 0 -2j], ["green", "red", "yellow", "blue"])
```

### Recursion

```python
def generate_points_diff(num, mul):
    mul -= 0.7
    result = []
    while num >= 1:
        calc_one = (num * mul, -num - 2)
        calc_two = (num * mul, num + 2)
        calc_three = (-num * mul, -num - 2)
        calc_four = (-num * mul, num + 2)
        result.append(calc_one)
        result.append(calc_two)
        result.append(calc_three)
        result.append(calc_four)
        num -= 1
    return result

def generate_points(num):
    for i in range(2, 6):
        result = generate_points_diff(num, i)
    while num >= 1:
        calc_one = (num * 1.5, -num - 2)
        calc_two = (num * 1.5, num + 2)
        calc_three = (-num * 1.5, -num - 2)
        calc_four = (-num * 1.5, num + 2)
        result.append(calc_one)
        result.append(calc_two)
        result.append(calc_three)
        result.append(calc_four)
        num -= 1
    return result
points_array = np.array(generate_points(10))
x = points_array[:, 0]
y = points_array[:, 1]
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
